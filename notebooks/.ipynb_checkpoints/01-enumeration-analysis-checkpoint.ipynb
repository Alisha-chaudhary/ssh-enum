```python
import sys; sys.path.insert(0, '..')
from src.attack_tools.manual_ssh import ManualSSHEnumerator

enumerator = ManualSSHEnumerator(target_ip='192.168.56.10')
results = enumerator.test_usernames(
    usernames=['root', 'admin', 'postgres', 'ubuntu', 'fakeuser999'],
    password='wrongpassword123'
)
```

```python
import pandas as pd
df = pd.DataFrame(results)
df['mean_ms'] = df['mean_s'] * 1000
print(df[['username','result','mean_ms']])
```

```python
import matplotlib.pyplot as plt
plt.bar(df['username'], df['mean_ms'])
plt.ylabel('Mean response time (ms)')
plt.title('SSH response times: valid vs invalid users')
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig('../screenshots/timing-comparison.png')
plt.show()
```
